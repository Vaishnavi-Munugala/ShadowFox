```python
import matplotlib.pyplot as plt
import numpy as np
days = np.arange(1, 8)
sales = [200, 220, 250, 230, 210, 190, 180]
plt.plot(days, sales, marker = 's', color = 'green')
plt.title("Line Plot - Sales over Days")
plt.xlabel("Day")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
students = [" John", "Emma", "Amit", "Sara"]
scores = [85, 92, 78, 90]
plt.bar(students, scores, color='orange')
plt.title("Student Test Scores")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [5, 12, 17, 23, 45, 44, 33, 21, 15, 10, 9, 8, 6, 2, 1]
plt.hist(data, bins=5, color='blue', edgecolor='black')
plt.title("Simple Histogram")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

age = np.random.randint(18, 30, 30)
score = np.random.randint(60, 100, 30)

plt.scatter(age, score, color='purple')
plt.title("Scatter Plot - Age vs Test Score")
plt.xlabel("Age (years)")
plt.ylabel("Test Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ["Math", "Science", "English", "History"]
marks = [30, 25, 20, 15]

plt.pie(marks, labels=subjects, autopct='%1.1f%%')
plt.title("Pie Chart - Subjects Marks Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 6]

plt.fill_between(x, y, color='orange')
plt.title("Area Chart using fill_between()")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 6]

plt.stackplot(x, y1, y2, labels=['Series 1', 'Series 2'], colors=['lightgreen', 'lightcoral'])
plt.title("Area Chart using stackplot()")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset("exercise")

sns.stripplot(x='time', y='pulse', hue='kind', data=df)
plt.title('Scatter Plot: Pulse vs Time, by Kind')
plt.xlabel('Time')
plt.ylabel('Pulse')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.lineplot(x='kind', y='pulse', hue='diet', data=df)
plt.title('Line Plot: Pulse by Kind and Diet')
plt.xlabel('Kind')
plt.ylabel('Pulse')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.barplot(x='kind', y='pulse', hue='diet', data=df)
plt.title('Bar Plot: Pulse by Kind and Diet')
plt.xlabel('Kind')
plt.ylabel('Pulse')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.countplot(x='kind', hue='diet', data=df)
plt.title('Count Plot: Kind by Diet')
plt.xlabel('Kind')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.boxplot(x='kind', y='pulse', hue='diet', data=df)
plt.title('Box Plot: Pulse by Kind and Diet')
plt.xlabel('Kind')
plt.ylabel('Pulse')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset("exercise")

sns.histplot(data=df, x='pulse', hue='kind', kde=False, multiple='stack', palette='Set2')
plt.title('Histogram: Pulse by Kind')
plt.xlabel('Pulse')
plt.ylabel('Count')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.kdeplot(data=df, x='pulse', hue='kind')
plt.title('KDE Plot: Pulse by Kind')
plt.xlabel('Pulse')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset("exercise")

exercise_encoded = pd.get_dummies(df, columns=['kind', 'diet', 'time'], drop_first=True)

correlation_matrix = exercise_encoded.corr()

plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap of Exercise Data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset("exercise")

sns.regplot(x=df.index, y='pulse', data=df)
plt.title("Regression Plot: Index vs Pulse")
plt.xlabel("Index")
plt.ylabel("Pulse")
plt.show()
```
